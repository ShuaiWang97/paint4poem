```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-darkgrid")
import seaborn as sns
from numpy import polyfit, poly1d
from collections import defaultdict
```

```python
data_caption = pd.read_csv("training.csv", index_col=None, nrows=300)
data_web = pd.read_csv("training_from_web.csv", index_col=None, nrows=300)
data_stratch = pd.read_csv("training_new_stratch.csv", index_col=None, nrows=700)
data_lambda_01 = pd.read_csv("training_lambda_0.1.csv", index_col=None, nrows=300)
data_lambda_10 = pd.read_csv("training_lambda_10.csv", index_col=None, nrows=300)

data_list = [data_caption, data_web,data_stratch,data_lambda_01,data_lambda_10 ]
data_name = ["data_caption", "data_web","data_stratch","data_lambda_01","data_lambda_10"]
```

```python
names =  locals()
for num, i in enumerate(data_list):
    names["errD_tot_"+data_name[num]]  = i["errD_total"]
    names["errD_cond_"+data_name[num]] = i["cond_errD_total"]
    names["errD_uncond_"+data_name[num] ]  = i["uncond_errD_total"]
    names["epoch_"+data_name[num]] = i["epoch"]
    names["errG_tot_"+data_name[num]]  = i["errG_total"]
    names["errG_cond_"+data_name[num] ]= i["cond_errG"]
    names["errG_uncond_"+data_name[num]]   = i["uncond_errG"]

    names["d_"+data_name[num]] = defaultdict()
    names["d_"+data_name[num]]["D"] = [names["errD_tot_"+data_name[num]] , names["errD_cond_"+data_name[num]], names["errD_uncond_"+data_name[num]]] 
    names["d_"+data_name[num]]["G"] = [names["errG_tot_"+data_name[num]] ,names["errG_cond_"+data_name[num] ],names["errG_uncond_"+data_name[num]]]
'''
errD_tot_web  = data_web["errD_total"]
errD_cond_web = data_web["cond_errD_total"]
errD_uncond_web   = data_web["uncond_errD_total"]
epoch_web = data_web["epoch"]
errG_tot_web  = data_web["errG_total"]
errG_cond_web = data_web["cond_errG"]
errG_uncond_web   = data_web["uncond_errG"]

d_web = defaultdict()
d_web["D"] = [errD_tot_web ,errD_cond_web,errD_uncond_web ]
d_web["G"] = [errG_tot_web ,errG_cond_web,errG_uncond_web ]
'''
```

```python
name_list = [names["d_"+i] for i in data_name]
for numm, ii in enumerate(name_list):
    for num, i in enumerate([ii["D"], ii["G"]]):
        if num == 0:
            name = "D"
        else:
            name = "G"
        fig, (axes1, axes2, axes3) = plt.subplots(1, 3, figsize=(15, 5))
        
        epoch = names["epoch_"+data_name[numm]]
        axes1.plot(epoch, i[0],label = "initial data")
        coeff1 = polyfit(epoch, i[0], 3)
        axes1.plot(epoch, coeff1[0] * epoch**3 + coeff1[1]*epoch*2 + coeff1[2]*epoch + coeff1[3], 'r-',label = "fitted curve")
        axes1.set_title(name + "_loss_total")
        axes1.set_xlabel("epoch")
        axes1.legend()
        #axes1.set_ylabel("Discriminator total loss")

        axes2.plot(epoch, i[1],label = "initial data")
        coeff2 = polyfit(epoch, i[1], 2)
        axes2.plot(epoch, coeff2[0] * epoch**2 + coeff2[1]*epoch + coeff2[2], 'r-',label = "fitted curve")
        axes2.set_title("err" + name + "_cond")
        axes2.set_xlabel("epoch")
        axes2.legend()
        #axes2.set_ylabel("Discriminator errD_cond")

        axes3.plot(epoch, i[2],label = "initial data")
        coeff3 = polyfit(epoch, i[2], 2)
        axes3.plot(epoch, coeff3[0] * epoch**2 + coeff3[1]*epoch + coeff3[2], 'r-',label = "fitted curve")
        axes3.set_title("err" + name + "_uncond")
        axes3.set_xlabel("epoch")
        axes3.legend()
        #axes3.set_ylabel("Discriminator uncond_errD")
```

```python
name_list = [names["d_"+i] for i in data_name]
for numm, ii in enumerate(name_list[0:3]):
    for num, i in enumerate([ii["D"], ii["G"]]):
        if num == 0:
            name = "D"
        else:
            name = "G"
        fig, (axes1, axes2, axes3) = plt.subplots(1, 3, figsize=(15, 5))
        
        epoch = names["epoch_"+data_name[numm]]
        axes1.plot(epoch, i[0],label = "initial data")
        coeff1 = polyfit(epoch, i[0], 2)
        axes1.plot(epoch, coeff1[0] * epoch**2 + coeff1[1]*epoch + coeff1[2], 'r-',label = "fitted curve")
        
        
        axes1.set_title(name + "_loss_total")
        axes1.set_xlabel("epoch")
        axes1.legend()
        #axes1.set_ylabel("Discriminator total loss")

        axes2.plot(epoch, i[1],label = "initial data")
        coeff2 = polyfit(epoch, i[1], 2)
        axes2.plot(epoch, coeff2[0] * epoch**2 + coeff2[1]*epoch + coeff2[2], 'r-',label = "fitted curve")
        axes2.set_title("err" + name + "_cond")
        axes2.set_xlabel("epoch")
        axes2.legend()
        #axes2.set_ylabel("Discriminator errD_cond")

        axes3.plot(epoch, i[2],label = "initial data")
        coeff3 = polyfit(epoch, i[2], 2)
        axes3.plot(epoch, coeff3[0] * epoch**2 + coeff3[1]*epoch + coeff3[2], 'r-',label = "fitted curve")
        axes3.set_title("err" + name + "_uncond")
        axes3.set_xlabel("epoch")
        axes3.legend()
        #axes3.set_ylabel("Discriminator uncond_errD")
```

```python
# data_list = [data_caption, data_web,data_stratch,data_lambda_01,data_lambda_10 ]
# ata_name = ["data_caption", "data_web","data_stratch","data_lambda_01","data_lambda_10"]

ana_list= defaultdict()
ana_list["D"] = ["errD_tot"  ,"errD_cond", "errD_uncond" ]
ana_list["G"] = ["errG_tot"  ,"errG_cond", "errG_uncond" ]
DG_LIST= [""] 
for numm, ii in enumerate([ana_list["D"], ana_list["G"]]):
    fig, (axes1, axes2, axes3) = plt.subplots(1, 3, figsize=(15, 5))
    for num, i in enumerate(ii):
        coeff11 = polyfit(epoch_data_caption, names[i + "_data_caption"], 3)
        coeff12 = polyfit(epoch_data_web, names[i + "_data_web"], 3)
        coeff13 = polyfit(epoch_data_stratch, names[i + "_data_stratch"], 3)
        epoch = epoch_data_caption
        names["axes" +str(num+1)].plot(epoch, coeff11[0] * epoch**3 + coeff11[1]*epoch**2 + coeff11[2]*epoch + coeff11[3], 'y-',label = "caption pretrain")
        epoch = epoch_data_web
        names["axes" +str(num+1)].plot(epoch, coeff12[0] * epoch**3 + coeff12[1]*epoch**2 + coeff12[2]*epoch + coeff12[3], 'b-',label = "web pretrain")
        epoch = epoch_data_stratch
        names["axes" +str(num+1)].plot(epoch, coeff13[0] * epoch**3 + coeff13[1]*epoch**2 + coeff13[2]*epoch + coeff13[3], 'r-',label = "from stratch")
        names["axes" +str(num+1)].legend()
        names["axes" +str(num+1)].set_title(i+"_loss")
```
